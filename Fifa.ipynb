```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.head()
```

### Histograms

```python
x=[40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=x)
plt.title('Players vs Rating', fontdict={'fontweight':'bold','fontsize':22})
plt.xlabel('Overall Rating')
plt.ylabel('No. Of Players')
plt.xticks(x)
plt.show()
```

### Pie Chart

```python
fifa.head(2)
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right= fifa.loc[fifa['Preferred Foot']== 'Right'].count()[0]

labels =['Left','Right']
colors = ['#00FFFF','#FF00FF']
plt.pie([left,right],labels=labels,colors=colors, autopct='%.2f %%')

plt.title('Foot Pref. Of Fifa Players')
plt.savefig('Foot Preference.png',dpi=100)

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)== str else x for x in fifa.Weight]
```

```python
light = fifa.loc[fifa.Weight<140].count()[0]
light_medium = fifa.loc[(fifa.Weight>140)&(fifa.Weight<160)].count()[0]
medium = fifa.loc[(fifa.Weight<180)&(fifa.Weight>160)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight<200)&(fifa.Weight>180)].count()[0]
heavy = fifa.loc[fifa.Weight>200].count()[0]

weights=[light,light_medium,medium,medium_heavy,heavy]

labels = ['<140','140-160','160-180','180-200','>200']
colors = ['r','b','y','#00FFFF','#FF00FF']
plt.pie(weights,labels = labels, colors=colors, autopct='%.2f %%')
plt.title('Fifa Players Weight Distribution')

plt.savefig('Fifa Player Weight Distribution.png',dpi=100)
plt.show()
```
